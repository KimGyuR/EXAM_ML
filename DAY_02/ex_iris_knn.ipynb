## 목표 : 붓꽃 품종 식별 모델
- 데이터 : 3개 품종을 가진 iris.csv
- 피 쳐 : sepal.length, sepal.width
- 타 겟 : variety
- 유 형 : 지도학습 / 분류

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
data_file = '../data/iris.csv'
```

```python
irisDF = pd.read_csv(data_file, usecols=[0,1,4])

irisDF.info()
```

```python
# Setosa, Virginica 품종 행만 추출
mask = irisDF.variety.isin(['Setosa', 'Virginica'])  # Boolean Indexing

# 해당 행만 도출
twoirisDF = irisDF[mask]
twoirisDF.index

# 인덱스 정리
twoirisDF.reset_index(drop=True, inplace=True)
twoirisDF
```

```python
## Setosa, Virginica 품종명을 수치화
twoirisDF.variety = twoirisDF.variety.replace({'Setosa':0, 'Versicolor':1, 'Virginica':2}).copy()

twoirisDF
```

```python
twoirisDF = twoirisDF.sample(frac=1.)

print(twoirisDF.shape)
twoirisDF
```
