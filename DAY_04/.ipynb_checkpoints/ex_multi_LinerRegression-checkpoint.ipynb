### 목표 : 다중 속성/피쳐를 기반으로 수치값 예측
- 데이터 : fish.csv
- 피쳐 : 4개 Length, Diagonal, Height, Width
- 타겟 : Weight
- 학습유형: 지도학습 + 회귀
- 알고리즘 : 데이터를 확인 후 결정

(1) 모듈로딩 및 데이터 준비 <hr>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
# 데이터 준비
data_file = '../data/fish.csv'

# Bream 행만 추출 => 35개 행만 추출
dataDF = pd.read_csv(data_file, nrows=35, usecols=[1,2,3,4,5])
```

```python
# 기본 정보
dataDF.info()
```

(2) 데이터 확인 <hr>

- 타겟과 피쳐 관계 => 상관계수, 시각화

```python
# 피쳐와 타겟과의 관계 정도를 시각화 및 수치값으로 표기 및 출력
# 매개변수 : 행수, 열수, 타겟, 피쳐리스트, 상관계수값
# 반환 : 없음
def print_feature(nrows, ncols, df, target, features, corrs):
    for idx in range(len(features)):
        plt.subplot(nrows, ncols, idx+1)
        plt.scatter(df[target], df[features[idx]], label=f'corr:{corrs[idx]:.2}')
        plt.legend()
        plt.xlabel(target)
        plt.ylabel(features[idx])
    plt.tight_layout()
    plt.show()
```

```python
dataDF.columns[1:]
```

```python
corrList = dataDF.corr()['Weight'].to_list()
corrList[1:]
```

```python
print_feature(2, 2, dataDF, dataDF.columns[0], dataDF.columns[1:], corrList[1:])
```

(3) 학습 준비 <hr>
- 학습 알고리즘 : 양의 선형관계 파악됨 => 선형 모델
- 학습 데이터 준비 : 피쳐-타겟 분리, 학습용 데이터셋 - 테스트용 데이터셋
- 피쳐 값의 범위 비교 : 작은 범위 최대값 8 ~ 큰 범위 최대값 50 => 피쳐 스케일링

(3-1) 학습 데이터 분리 => 피쳐-타겟, 학습용 데이터셋-테스트용 데이터셋

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
# 피쳐와 타겟 분리
featureDF=dataDF[dataDF.columns[1:]]
featureDF.head(2)
```

```python
targetDF=dataDF[dataDF.columns[0]]
targetDF.head(2)
```

```python
# 데이터셋 준비 => 70:30
X_train, X_test, y_train, y_test = train_test_split(featureDF, 
                                                    targetDF, 
                                                    random_state=7, 
                                                    test_size=0.3)
```

```python
print(f'[훈련용] X_train : {X_train.shape} y_train : {y_train.shape}')
print(f'[테스트용] X_test : {X_test.shape} y_test : {y_test.shape}')
```

(3-2) 피쳐 스케일링 ==> 훈련데이터로 생성

```python
std_scaler = StandardScaler()
std_scaler.fit(X_train)
```

```python
mm_scaler = MinMaxScaler()
mm_scaler.fit(X_train)
```

```python
rb_scaler = RobustScaler()
rb_scaler.fit(X_train)
```

```python
# 3가지 버전에 스케일된 데이터셋 준비
X_train_std = std_scaler.transform(X_train)
X_test_std = std_scaler.transform(X_test)

X_train_mm = mm_scaler.transform(X_train)
X_test_mm = mm_scaler.transform(X_test)

X_train_rb = rb_scaler.transform(X_train)
X_test_rb = rb_scaler.transform(X_test)
```

(4) 학습 진행 <hr>

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train_std, y_train)

print(f'[모델 파라미터] ====>\ncoef_: {model.coef_}\nmodel.intercept_: {model.intercept_}')
```

(5) 모델 평가 <hr>
- 훈련용 데이터셋의 점수와 테스트 데이터셋 점수 비교
    * 훈련 점수 > 테스트 점수 : 과대적합
    * 훈련 점수 ▼, 테스트 점수 ▼ : 과소적합
    * 훈련 점수 ≒ 테스트 점수 : 최적적합

```python
# StandardScaler: 평균0, 편차1
def tt_std(X_train_std, y_train, X_test_std, y_test):
    train_score_std = model.score(X_train_std, y_train)
    test_score_std = model.score(X_test_std, y_test)
    
    print(f'Train Score std : {train_score_std} --- Test Score std : {test_score_std}')
    return train_score_std, test_score_std
```

```python
a,b= tt_std(X_train_std, y_train, X_test_std, y_test)
```

```python
plt.plot(X_train_std, y_train, X_test_std, y_test)
```

```python
# MinMaxScaler: 최소 0 최대 1
def tt_mm(X_train_mm, y_train, X_test_mm, y_test):
    train_score_mm = model.score(X_train_mm, y_train)
    test_score_mm = model.score(X_test_mm, y_test)

    print(f'Train Score mm : {train_score_mm} --- Test Score mm : {test_score_mm}')
    return train_score_mm, test_score_mm
```

```python
c,d= tt_mm(X_train_mm, y_train, X_test_mm, y_test)
```

```python
plt.plot(X_train_mm, y_train, X_test_mm, y_test)
```

```python
# RobustScaler: 평균0, 편차1
def tt_rb(X_train_rb, y_train, X_test_rb, y_test):
    train_score_rb = model.score(X_train_rb, y_train)
    test_score_rb = model.score(X_test_rb, y_test)
    
    print(f'Train Score rb : {train_score_rb} --- Test Score rb : {test_score_rb}')
    return train_score_rb, test_score_rb
```

```python
e,f=tt_rb(X_train_rb, y_train, X_test_rb, y_test)
```

```python
plt.plot(X_train_rb, y_train, X_test_rb, y_test)
```

```python
a,b,c,d,e,f
```

```python
plt.plot(['standard', 'minmax', 'robust'], [b, d, f], 'go-', label='test_score', linewidth=2)
plt.plot(['standard', 'minmax', 'robust'], [a, c, e], 'bo-', label='train_score', linewidth=2)
plt.legend()
plt.show()
```
