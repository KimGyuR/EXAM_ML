```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
```

```python
irisDF = pd.read_csv('../data/iris.csv')
irisDF = irisDF.iloc[:100]
irisDF
```

```python
featureDF = irisDF[irisDF.columns[:2]]
targetSR = irisDF[irisDF.columns[4]]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(featureDF,
                                                   targetSR,
                                                   stratify=targetSR,
                                                   test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler

sdScaler = StandardScaler()
sdScaler.fit(X_train)
```

```python
# 훈련용 데이터셋 스케일링
scaled_X_train=sdScaler.transform(X_train)

# 테스트용 데이터셋 스케일링
scaled_X_test = sdScaler.transform(X_test)
```

```python
# 위에서 standardscaler로 정규화 후 밑에 SVC모델로 분류
```

```python
X = np.array(featureDF)
y = targetSR.T.replace({"Setosa":0, "Versicolor":1})
print(f'{X.shape}, {y.shape}')
# SVC 모델 생성
svc_model = SVC(kernel='linear')
svc_model.fit(X, y)
```

```python
# 결정 경계 그리기
w = svc_model.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(X[:, 0].min(), X[:, 0].max())
yy = a * xx - (svc_model.intercept_[0]) / w[1]
```

```python
# Support vectors
margin = 1 / np.sqrt(np.sum(svc_model.coef_ ** 2)) #분산 
yy_down = yy - np.sqrt(1 + a ** 2) * margin
yy_up = yy + np.sqrt(1 + a ** 2) * margin

plt.plot(xx, yy, 'k-')
plt.plot(xx, yy_down, 'k--') # 점선
plt.plot(xx, yy_up, 'k--') # 점선 

# Scatter plot
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
# plt.ylim(-0.5,2)
plt.xlabel('sepal.length')
plt.ylabel('sepal.width')
plt.title('SVC Decision Boundary')
plt.show()
```
